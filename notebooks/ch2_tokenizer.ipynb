```python
import re

text_path = project_root / "data/raw/the-verdict.txt"

with open(text_path, "r", encoding="utf-8") as f:
    raw_text = f.read()

print(raw_text[:100])
```

```python
text = "Hello world, this is a test!"

result = re.split(r'([,.:;?_!"()\']|--|\s)', text)
result = [item.strip() for item in result if item.strip()]

print(result)
```

```python
preprocessed = re.split(r"([,.:?_!\"()\']|--|\s)", raw_text)
preprocessed = [item.strip() for item in preprocessed if item.strip()]
print("token length in text:", len(preprocessed))

all_words = sorted(set(preprocessed))
vocab_size = len(all_words)
print("vocab size:", vocab_size)

vocab = {token:integer for integer, token in enumerate(all_words)}
print(vocab)
```

```python
inverse_dict = {s:i for i,s in vocab.items()}

print(inverse_dict)
```

```python
from src.layers.simple_tokenizers import SimpleTokenizerV1

tokenizer = SimpleTokenizerV1(vocab)
text = """"It's the last he painted, you know,"
Mrs. Gisburn said with pardonable pride."""

tokens = tokenizer.encode(text)

print("tokens:", tokens)
print("decoded text:", tokenizer.decode(tokens))
```

**Adding tokens to account for unkown words and endoftext**

```python
all_tokens = sorted(list(set(preprocessed)))
all_tokens.extend(["<|endoftext|>", "<|unk|>"])
vocab = {token:integer for integer,token in enumerate(all_tokens)}

print(vocab)
print("new vocab lenghth:", len(vocab))
```

```python
from src.layers.simple_tokenizers import SimpleTokenizerV2

tokenizer = SimpleTokenizerV2(vocab)
text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the palace."
text = " <|endoftext|> ".join((text1, text2))
print(text)

tokens = tokenizer.encode(text)
print("tokens:", tokens)
print("decoded text:", tokenizer.decode(tokens))
```

**Byte-pair encoding**

```python
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")

text = (
"Hello, do you like tea? <|endoftext|> In the sunlit terraces"
"of someunknownPlace."
)
integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(integers)

strings = tokenizer.decode(integers)
print(strings)
```

```python
# Tiktoken BPE tokenizer handles unkown words by breaking down into known tokens (i.e. letters, common letter groups)
text = "Akwirw ier"
integers = tokenizer.encode(text)
print(integers)

for integer in integers:
    print(integer, "-", tokenizer.decode([integer]))
```

**Data sampling with a sliding window**

We need to generate the input-target pairs required for LLM training

```python
text_path = project_root / "data/raw/the-verdict.txt"

with open(text_path, "r", encoding="utf-8") as f:
    raw_text = f.read()

enc_text = tokenizer.encode(raw_text)
print("encoded text length:", len(enc_text))
```

```python
enc_sample = enc_text[:50]

context_size = 4
print(enc_sample)
x = enc_sample[:context_size]
y = enc_sample[1:context_size+1]
print(f"x: {x}")
print(f"y:     {y}")
```

```python
for i in range(1, context_size+1):
    x = enc_sample[:i]
    y = enc_sample[i]
    print(x, "-->", y)
    print(tokenizer.decode(x), "-->", tokenizer.decode([y]), "\n")
```

**Encoding word positions**

Lets consider GPT-3's vocab_size and output_dim:

```python
import torch
vocab_size = 50257
output_dim = 256

token_embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
print(token_embedding_layer.weight)
```

```python
max_length = 4
dataloader = create_dataloader_v1(
    raw_text, batch_size=8, max_length=max_length,
    stride=max_length, shuffle=False
    )
data_iter = iter(dataloader)
inputs, targets = next(data_iter)
print("Token IDs:\n", inputs)
print("\nInputs shape:", inputs.shape)
```

```python
token_embeddings = token_embedding_layer(inputs)
print(token_embeddings.shape)
```

```python
context_length = max_length
pos_embedding_layer = torch.nn.Embedding(context_length, output_dim)
pos_embeddings = pos_embedding_layer(torch.arange(context_length))
print(pos_embeddings.shape)
```

```python
input_embeddings = token_embeddings + pos_embeddings
print(input_embeddings.shape)
```
